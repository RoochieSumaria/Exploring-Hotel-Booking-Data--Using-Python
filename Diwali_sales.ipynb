```python
ax= sns.countplot(x='Gender',data=df)

for i in ax.containers:                          # ax.containers have value of bar column using for loop display bar_label
    ax.bar_label(i)
```

```python
Female have are count means they spent more in Diwali Shopping
```

```python
gp=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
ax=sns.barplot(x='Gender',y='Amount',data=gp)

                     #Display labels on the bar
```

```python
df.columns
```

```python
ax= sns.countplot(data=df, x='Age Group', palette='Set3', hue='Gender')

for i in ax.containers:
    ax.bar_label(i)
```

Most of the buyers are from Age group 26-35 years old.

```python
df.columns
```

```python
#Total number of orders from top 10 states
states_gp=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(x='State', y='Orders',data=states_gp, palette='Set1')
```

From above graphs we can see that most of orders are from Uttar Pradesh, Maharashtra and Karnataka respectively

```python
#Total number of states/sales from top 10 states
states_gp=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(x='State', y='Amount',data=states_gp, palette='Set1')
```

From above graphs we can see that most of amount spending are from Uttar Pradesh, Maharashtra and Karnataka respectively

Marital status

```python
ax= sns.countplot(data=df, x='Marital_Status')
sns.set(rc={'figure.figsize':(15,5)})
for i in ax.containers:
    ax.bar_label(i)
```

```python
M=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=M, x='Marital_Status', y='Amount', hue='Gender')
```

Married Women have spent more amount in purchasing

Occupation

```python
ax= sns.countplot(data=df, x='Occupation',palette='Set2')

for i in ax.containers:
    ax.bar_label(i)
```

```python
O=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=O, x='Occupation', y='Amount', palette='Set2')
```

For above graphs we can see that buyers are from IT, Healthcare and Aviation sectors.

Product Category

```python
sns.set(rc={'figure.figsize':(30,3)})
sns.countplot(data=df,x='Product_Category')
```
