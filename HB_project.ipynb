```python
plt.title('Reservation in both hotels in the years(2015,2016,2017)')

sns.countplot(data=HBooking_data, x='hotel', hue='arrival_date_year', palette='Set3')

plt.show() #is_repeated_guest
```

```python
percentage = HBooking_data['is_canceled'].value_counts(normalize=True) * 100

sns.barplot(x=percentage.index, y=percentage, palette=['brown', 'grey'])
plt.title('Percentage of Reservations & Cancellations')
plt.xlabel('Is Cancelled')
plt.ylabel('Percentage (%)')
plt.xticks(ticks=[0, 1], labels=['Not Cancelled', 'Cancelled'])
plt.show()
```

```python
city = HBooking_data[HBooking_data['hotel'] == 'City Hotel']
```

```python
resort = HBooking_data[HBooking_data['hotel'] == 'Resort Hotel']
```

```python
resort = resort.groupby('reservation_status_date')[['adr']].mean()
city = city.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (10, 5))
sns.lineplot(data = resort, x = 'reservation_status_date', y = 'adr', label = 'Resort Hotel')
sns.lineplot(data = city, x = 'reservation_status_date', y = 'adr', label = 'City Hotel')

plt.ylabel('Average Daily Rate (ADR)')

plt.title('Average Daily Rate (ADR) by CityHotel Vs ResortHotel')
plt.legend(fontsize = 10)
plt.show()
```

```python
resort_cancelled = resort[(resort['is_canceled'] == 1) & (resort['reservation_status_date'] > '2017')]
resort_not_cancelled = resort[(resort['is_canceled'] == 0) & (resort['reservation_status_date']> '2017')]
```

```python
plt.figure(figsize=(10, 5))
ax2 = sns.lineplot(data=resort_cancelled, x='reservation_status_date', y='lead_time', label='Cancelled')
sns.lineplot(data=resort_not_cancelled, x='reservation_status_date', y='lead_time', label='Not Cancelled', ax=ax2)
plt.title('Reservation status with lead time for Resort Hotel', fontsize = 15)
plt.ylabel('Lead time')
plt.legend(fontsize=10)
plt.show()
```

Cancellation due to more lead time for Resort Hotel
